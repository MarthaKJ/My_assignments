## SERIES

```python
# 1) compare the elements of two panda series
import pandas as pd
series1 = pd.Series([4,65,436,3,9])
series2 = pd.Series([7,0,3,897,9])
comparision = series1==series2
print(comparision)
```

```python
#2) add , subtract, multiply and divide two panda series
series1 = pd.Series([2,4,6,8,14])
series2 = pd.Series([1,3,5,7,9])

#addition
addition = series1 + series2
print(f"Addition:{addition}")

#subtraction
subtraction = series1 - series2
print(f"Subtraction : {subtraction}")

#Multiplication
multiplication = series1*series2
print(f"Multipliction: {multiplication}")

#Division 
division = series1/series2
print(f"Division: {division}")
```

```python
#3) convert a dictionary to a pandas series
dictionary1 = {'Josh':24, 'Sam':36 , 'Peaace':19, 'Charles': 65 , 'Tom':44}
series = pd.Series(dictionary1)
print(series)
```

```python
#4) to convert a given series to an array
import numpy as np
series = pd.Series(['Love',800 , 'Joy' , 789.9 , 'Peace', True])
array = np.array(series)
print(array)
```

```python
#5) display the mostfrequent value in a given series
#and replace everything else as 'Other' in the series

df = pd.read_csv("C:/Users/hp/Desktop/vE/mienv/AfricaCupofNationsMatches.csv")
most_frequent = df['HomeTeamGoals'].value_counts().idxmax()

df['HomeTeamGoals']= df['HomeTeamGoals'].apply(lambda x: 'Other' if x!= most_frequent else x)
print(df['HomeTeamGoals'])
```
